### Handling missing values
- Missing data is usually either not present (empty string) or marked by some *sentinel* value. 
- By default, pandas uses a set of commonly occurring sentinels, such as NA and NULL:

```python
!cat examples/ex5.csv
```

```python
result = pd.read_csv('examples/ex5.csv')
result  ## missing values are auto-detected
```

"na_values" option can take either a list or set of strings to consider missing values

```python
result = pd.read_csv('examples/ex5.csv', \
                     na_values=['foo']) 
result
```

## Iterating: Reading Text Files in Pieces

When processing very large files or figuring out the right set of arguments to correctly process a large file, you may only want to read in a small piece of a file or iterate through smaller chunks of the file.

```python
# let's make pandas display settings more compact
pd.options.display.max_rows = 10  

result = pd.read_csv('examples/ex6.csv')
result
```

```python
# to read a small number of rows (avoiding reading the entire file), specify that with nrows:
pd.read_csv('examples/ex6.csv', nrows=5)
```

## Example (column average)

```python
tst=pd.read_csv('examples/ex6.csv')
tst['one'].agg('mean')
```

```python
chunker = pd.read_csv('examples/ex6.csv', chunksize=100)
```

```python
one_average=[]
for chunk in chunker:
    chunk_ave=chunk['one'].agg(['mean','count']).tolist()
    one_average.append(chunk_ave)
```

```python
rs1=pd.DataFrame(one_average)
```

```python
sum(rs1[0]*rs1[1])/ sum(rs1[1])
```

## When all fails - reading manually

```python
! cat 'examples/ex7.csv'
```

```python
import re
def line_reader(line, sep=',',eol='\n'):
    row=line.split(sep)
    row=[cell.replace('\"','') for cell in row]
    row[-1]=row[-1].replace(eol,'')
    return row
```

```python
file=open('examples/ex7.csv', )
rows=[]
for line in  file.readlines():
    rows.append(line_reader(line))
```

```python
pd.DataFrame(rows[1:], columns=rows[0])
```

## JSON (JavaScript Object Notation) Data
- JSON has become one of the standard formats for sending data by HTTP request between web browsers and other applications.
- It is a much more free-form data format than a tabular text form like CSV.
- Almost python valid

```python
obj = """
{"name": "Wes",
 "places_lived": ["United States", "Spain", "Germany"],
 "pet": null,
 "siblings": [{"name": "Scott", "age": 30, "pets": ["Zeus", "Zuko"]},
              {"name": "Katie", "age": 38,
               "pets": ["Sixes", "Stache", "Cisco"]}]
}
"""
```

```python
import json 
result = json.loads(obj)
result
```

```python
result['siblings']
siblings = pd.DataFrame(result['siblings'], columns=['name', 'age'])
siblings
```

## Binary Data Formats
- One of the easiest ways to store data (also known as serialization) efficiently in binary format is using Pythons built-in pickle serialization
- Pickle is only recommended as a short-term storage format (version dependency).

```python
frame = pd.read_csv('examples/ex1.csv')
frame
frame.to_pickle('examples/frame_pickle')
pd.read_pickle('examples/frame_pickle')
```
