```python
import requests
url_scam='http://www.419scam.org/emails'
page=requests.get(url_scam)
page
```

```python
from bs4 import BeautifulSoup
```

```python
url_text=page.text
print(url_text)
```

```python
import re
```

```python

url_tags=re.findall(r'<.*?>',url_text)
url_tags
```

```python
set(url_tags)
```

```python
url_links=re.findall(r'<a .*/a>',url_text)
```

```python
url_links
```

```python
def get_text(source):
    href_text_regex=re.compile(r'>.*<')
    href=re.findall(href_text_regex, source)
    
    if len(href)>0:
        clean_text = re.compile('<.*?>')
        return re.sub(clean_text, '', href[0][1:-1])
    else: 
        return ''

def get_href(source):
    href_regex= re.compile('href=".*?"')
    href=re.findall(href_regex, source)
    #print(href)
    if len(href)>0:
        return  href[0][6:-1]
    else:
        return ''
    
def get_date(source):
    href_text_regex=re.compile(r'\d{4}.*?[/]\d{2}')
    href=re.findall(href_text_regex, source)
    #print(href)
    if len(href)>0:
        return href[0]
    else: 
        return ''
```

```python
get_list =[ (get_text(link),get_href(link),get_date(link) ) for link in url_links]
print(get_list)
```

```python
import pandas as pd
```

```python
pd.DataFrame(get_list, columns=['Link Text','URL','Date'])
```
