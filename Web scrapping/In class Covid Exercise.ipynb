```python
days=[]
for i in range(1,30):
    days.append(date.today()- timedelta(days=i))
```

```python
days
```

```python
url='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/{}.csv'
```

```python
days[0].strftime('%m-%d-%Y')
```

```python
all_days=[]
for day in days: 
    day_str=day.strftime('%m-%d-%Y')
    print(url.format(day_str))
    day_data=pd.read_csv(url.format(day_str))
    day_data['day']=day_str
    all_days.append(day_data)
```

```python
all_day_dfs=pd.concat(all_days)
```

```python
all_day_dfs
```

```python
all_day_dfs[all_day_dfs['Province_State']=='Alabama']
```

```python
wide_table=all_day_dfs.pivot_table(index='Province_State', columns='day', values='Deaths')
```

```python
wide_table
```

```python
wide_table_rindex=wide_table.reset_index()

```
