## <font color='red'>Reminder:</font>  deciphering URL information  

- **The base** URL represents the path to the search functionality of the website. 

- **The query parameters** represent additional values that can be declared on the page.

### Find Elements by ID

```python
url_monster='https://www.monster.com/jobs/search/?q=Data-Scientist&where=NYC'
page=requests.get(url_monster)
soup = BeautifulSoup(page.content, 'html.parser')
results=soup.find(id='ResultsContainer')
print(results.prettify()[0:500])
```

### Find Elements by HTML Class Name

```python
job_elems = results.find_all('section', class_='card-content')
```

```python
for job_elem in job_elems:
    print(job_elem, end='\n'*2)
```

## Iterate Through Elements

```python
for job_elem in job_elems:
    # Each job_elem is a new BeautifulSoup object.
    # You can use the same methods on it as you did before.
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    print('####Title Element#########')
    print(title_elem)
    print('####Company Element#########')
    print(company_elem)
    print('####Location Element#########')
    print(location_elem)
    print()
    
```

### Extract Text From HTML Elements

```python
for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    print(title_elem.text)
    print(company_elem.text)
    print(location_elem.text)
    print()
```

### Handle Exceptions 

```python
for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        continue
    print(title_elem.text.strip())
    print(company_elem.text.strip())
    print(location_elem.text.strip())
    print()
```

### Compile to a Dataset

```python
listings=[]
for job_elem in job_elems:
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        continue
    listings.append([title_elem.text.strip(), company_elem.text.strip(), location_elem.text.strip()])
```

```python
import pandas as pd
```

```python
pd.DataFrame(listings, columns=['title','company','location'])
```
