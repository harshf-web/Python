```python
import requests
url = 'http://www.419scam.org/emails/'
page = requests.get(url)
page
```

```python
html_text=page.text
print(html_text)
```

```python
import re
all_links=re.findall(r'<a.*\d{4}-\d{2}/\d{0,2}.*/a>',html_text)
```

```python
all_links
```

```python
selected_links=list(range(0))
selected_links.insert(0,all_links[20])
selected_links.insert(2,all_links[10])
print(selected_links)
```

```python
all_selected_data_listing =list(range(0))
for link in selected_links:
    current_link=re.split('\"',link)
    url_selected_link=url+current_link[1]
    #print(url_selected_link)
    page1=requests.get(url_selected_link)
    html_text1=page1.text
    #print(html_text1)
    all_selected_links=re.findall(r'<a.*\d{4,}[.]\d.*/a>',html_text1)
    #print(all_selected_links)
    for link in all_selected_links:
        selected_data= re.sub(r'(<a\shref=\")', r'\1'+url_selected_link.replace("index.htm","")+'', link)
        #print(selected_data)
        all_selected_data_listing.append(selected_data)
        
print(all_selected_data_listing)
```

```python
def get_text(source):
    href_text_regex=re.compile(r'>.*<')
    href=re.findall(href_text_regex, source)
    #print(href)
    if len(href)>0:
        clean_text = re.compile('<.*?>')
        return re.sub(clean_text, '', href[0][1:-1])
    else: 
        return ''
def get_href(source):
    href_regex= re.compile('href=".*?"')
    href=re.findall(href_regex, source)
    if len(href)>0:
        return  href[0][6:-1]
    else:
        return ''
def get_date(source):
    href_text_regex=re.compile(r'\d{4}.*?[/]\d{2}')
    href=re.findall(href_text_regex, source)
    #print(href)
    if len(href)>0:
        return href[0]
    else: 
        return ''
```

```python
get_list =[ (get_text(item),get_href(item),get_date(item)) for item in all_selected_data_listing]
print(get_list)
```
