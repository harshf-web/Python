```python
all_the_articles = []
condition=True

while condition:
    article=browser.find_elements_by_tag_name('li[class="arxiv-result"]')
    for a_article in article:
        arxiv_id = a_article.find_elements_by_tag_name('a')[0].text
        pdf = a_article.find_elements_by_tag_name('a')[1].get_attribute('href')
        title = a_article.find_element_by_tag_name('p[class="title is-5 mathjax"]').text
        
        authors = a_article.find_element_by_tag_name('p[class="authors"]').text
        abstract = a_article.find_element_by_tag_name('p[class="abstract mathjax"]').text
        submitted =a_article.find_element_by_tag_name('p[class="is-size-7"]').text
    
        result = [arxiv_id,pdf,title,authors,abstract,submitted]
        all_the_articles.append(result)
    try:
        browser.find_element_by_tag_name('a[class="pagination-next"]').click()
    except:
        condition=False
```

```python
all_articles_final=pd.DataFrame(all_the_articles)
```

```python
all_articles_final.columns=['arxiv_id','pdf_link','title','authors','abstract','submitted']
```

```python
#cleaning
import nltk
import re
#nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')   ## Our stop words
wn = nltk.WordNetLemmatizer()
sm = nltk.PorterStemmer()

def clean(text):
    text=re.sub(r'[^\w\s]',' ',text).lower()   ### Remove punctuation by removing non-space alpha-num characters
    words=re.split(r'\W+', text)  ### Split by remaining non-alphanumeric characters 
    non_stop_words=[word for word in words if word not in stopword]        ### Remove stopwords
    lem_words= [sm.stem(word) for word in non_stop_words] 
    clean_text= ' '.join(lem_words)
    return clean_text
```

```python
search_abstract=search_abstract.apply(clean)
```

```python
search_abstract
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer=CountVectorizer()

vsm=count_vectorizer.fit_transform(search_abstract)
pd.DataFrame(vsm.todense(), columns=count_vectorizer.get_feature_names())

vocab=count_vectorizer.get_feature_names()
vocab
```

```python
query=['natural language processing']

## Vectorize the query
query_vector=count_vectorizer.transform(query)

query_vector.todense()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```
